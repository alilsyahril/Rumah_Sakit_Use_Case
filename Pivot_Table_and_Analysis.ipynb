### Pivot Table Using Pandas


```python
import pandas as pd
```

```python
df = pd.read_csv('Online Assesment - Business Data Support - Nur Syahril - Pelayanan Harian.csv')
df
```

```python
columns_to_drop = ['Unnamed: 4', 'Unnamed: 5']
df = df.drop(columns_to_drop, axis=1)
```

```python
df
```

```python
df_rabu = df[df['Hari'] == 'Wednesday']
df_rabu
```

```python
pivot_table = pd.pivot_table(data=df_rabu,
                             values='Jumlah Kunjungan',
                             index='Tanggal Jadwal Kunjungan',
                             columns='Poli',
                             aggfunc='sum')
```

```python
pivot_table
```

```python
# save into csv
pivot_table.to_csv('pivot_table.csv', index=True)
```
